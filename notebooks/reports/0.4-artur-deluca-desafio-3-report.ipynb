### Strategy

The strategy used in this notebook is to segment the dataset using the `k_means` algorithm and the `math test codes` present in the dataset. To predict the answers a Markov model is used for each of the segments generated.

An important aspect of the Markov Chain is its **order**. It basically define how many elements in the past are analyzed to draw the conclusion of which are the next elements. If the order is too low, the results may seem random. However if the order is too high, the Markov Chain becomes too conditioned to the training dataset, causing an overfit
